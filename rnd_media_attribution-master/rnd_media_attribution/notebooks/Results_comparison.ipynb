```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pickle
import os
plt.style.use('seaborn-dark')
```

```python
base_dir='../../ml/model/attention_model/'
list_dir_to_compare=[
    'media-attribution-de10f69-2020-10-30-13-49-31/attention_model/',  #  fb - covid
    'media-attribution-297128c-2020-11-03-17-06-43/attention_model/',  #  gg - aug
    'media-attribution-fb7414f-2020-11-03-19-03-51/attention_model/',  #  rtb - aug 
]
df_tfb=pd.read_csv(os.path.join(base_dir,list_dir_to_compare[0],'df_tri.csv')).set_index('trinomials')
df_tgg=pd.read_csv(os.path.join(base_dir,list_dir_to_compare[1],'df_tri.csv')).set_index('trinomials')
df_trtb=pd.read_csv(os.path.join(base_dir,list_dir_to_compare[2],'df_tri.csv')).set_index('trinomials')
```

```python
df_tri=pd.DataFrame()
df_tri['fb']=df_tfb['norm_att']
df_tri['gg']=df_tgg['norm_att']
df_tri['rtb']=df_trtb['norm_att']
df_tri['tot']=df_tri.sum(axis=1)
df_tri['fb_pb']=df_tfb['position_based']
df_tri['gg_pb']=df_tgg['position_based']
df_tri['rtb_pb']=df_trtb['position_based']
df_tri=df_tri.fillna(0)
df_tri
```

```python
base_dir='../../ml/model/attention_model/'
list_dir_to_compare=[
    'media-attribution-de10f69-2020-10-30-13-49-31/attention_model/',  #  fb - covid
    'media-attribution-297128c-2020-11-03-17-06-43/attention_model/',  #  gg - aug
    'media-attribution-fb7414f-2020-11-03-19-03-51/attention_model/',  #  rtb - aug 
]
df_tfb=pd.read_csv(os.path.join(base_dir,list_dir_to_compare[0],'df_tri.csv')).set_index('trinomials')
df_tgg=pd.read_csv(os.path.join(base_dir,list_dir_to_compare[1],'df_tri.csv')).set_index('trinomials')
df_trtb=pd.read_csv(os.path.join(base_dir,list_dir_to_compare[2],'df_tri.csv')).set_index('trinomials')

df_tpfb=pd.read_csv(os.path.join(base_dir,list_dir_to_compare[0],'df_tripos.csv')).set_index('trinomials')
df_tpgg=pd.read_csv(os.path.join(base_dir,list_dir_to_compare[1],'df_tripos.csv')).set_index('trinomials')
df_tprtb=pd.read_csv(os.path.join(base_dir,list_dir_to_compare[2],'df_tripos.csv')).set_index('trinomials')
df_tpfb
```

```python
df_tpfb = (df_tpfb.drop(list(df_tpfb.columns)[11:],axis=1).join(df_tfb[['norm_att', 'first_touch', 'last_touch', 'position_based']])
     .rename(columns={'norm_att': 'dl_weights'}).sort_values(by='dl_weights',ascending=False)
)
df_tpgg = (df_tpgg.drop(list(df_tpgg.columns)[7:],axis=1).join(df_tgg[['norm_att', 'first_touch', 'last_touch', 'position_based']])
     .rename(columns={'norm_att': 'dl_weights'}).sort_values(by='dl_weights',ascending=False)
)
df_tprtb = (df_tprtb.drop(list(df_tprtb.columns)[11:],axis=1).join(df_trtb[['norm_att', 'first_touch', 'last_touch', 'position_based']])
     .rename(columns={'norm_att': 'dl_weights'}).sort_values(by='dl_weights',ascending=False)
)
```

```python
df_tprtb
```

```python
df_tripos=pd.DataFrame()
dataset={'fb' : df_tpfb,'gg' : df_tpgg,'rtb' : df_tprtb}

for ii in range(0,3):
    for col,ds in dataset.items():
        df_tripos[str(ii) + '_' + col] = ds[str(ii)]
        df_tripos [str(ii) + '_w_' + col] = ds['dl_weights']
    df_tripos [str(ii) + '_sum'] = df_tripos[df_tripos.columns[df_tripos.columns.str.contains(str(ii)+'_w_')]].sum(axis=1)
    df_tripos = df_tripos.drop(list(df_tripos.columns[df_tripos.columns.str.contains(str(ii)+'_w_')]),axis=1)
df_tripos = df_tripos.fillna(0)
df_tripos = df_tripos.reset_index()
df_tripos['tri'] = df_tripos['trinomials'].str.replace('_nokwDft','').str.replace('_kwDft','')
df_tripos=df_tripos.drop(columns=['trinomials'],axis=1).set_index('tri')
df_tripos=df_tripos.groupby('tri').mean()
df_tripos
```

```python
df_tripos['sum']=df_tripos[['0_sum','1_sum','2_sum']].sum(axis=1)
df_tripos=df_tripos.sort_values(by = 'sum', ascending=False)
df_tripos.head(10)
```
